```python
from dotenv import load_dotenv

load_dotenv()   # 경로지정
```

```python
from langchain_openai import ChatOpenAI

# ChatOpenAI 안에 들어가서 보기
# ChatOpenAI 가 기본적으로 채택하는 모델 -> gpt-3.5-turbo


# OPENAI_API_KEY -> 자동으로 들어가도록 설정되어있음 (OPENAI_API_KEY이름이 아니라면, 직접 명시)
# OPENAI_KEY  -> 이름을 못찾음
# llm = ChatOpenAI() 


import os
api_key = os.getenv("OPENAI_KEY")  # 직접 불러와서
llm = ChatOpenAI(api_key=api_key)  # 지정 
```

- 노트북 -> 재사용성이 높은 함수를 만들어서 활용하는게 아님
- 하나의 노트북이 하나의 애플리케이션처럼 모든 코드를 가지고 있어야 함 -> 그렇기 때문에, 환경변수를 OPENAI_KEY로 한다면, 모든 노트북 파일에서 일일히 지정해줘야 하는 문제점이 생김

```python
ai_message = llm.invoke("AI강의로 어떤 강의가 있나요?")   # 실행하는 단계
```

```python
# AIMessage 라는 클래스 안에 content가 있는 형식
ai_message.content
```
